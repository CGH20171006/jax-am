```python
!nvidia-smi
import jax.numpy as np
import os
from cfd_am import *
import time
os.environ["CUDA_VISIBLE_DEVICES"]="0"
jax.config.update("jax_enable_x64", True)
%matplotlib notebook
```

```python
mesh = mesh3d([4e-3, 1e-3, 5e-4], [400,100,100])
mesh_local = mesh3d([3e-3, 5e-4, 2e-4], [300,50,40])

cell_conn_local = np.copy(mesh_local.cell_conn)
params = {}
params['mesh'] = mesh
params['mesh_local'] = mesh_local
params['rho'] = 8440.
params['cp'] = lambda T: (0.2441*np.clip(T,300,1563)+338.39) 
params['k'] = lambda T: 0.0163105*np.clip(T,300,1563)+4.5847 
params['visco'] = 7e-3
params['Marangoni'] = -2e-5
params['dt'] = 5e-6
params['t_OFF'] = 2.5e-3
params['speed'] = .8
params['X0'] = [1e-3,5e-4] 
params['rb'] = 5e-5
params['eta'] = 0.43
params['P'] = 195.
params['T_ref'] = 300.
params['Ts'] = 1563.
params['Tl'] = 1623.
example = AM_3d(params)
```

```python
start_time = time.time()
for i in range(0,800):
    example.time_integration()
    elapsed = time.time() - start_time

    print(f'time:{elapsed},T_max:{example.T0.max()},vmax:{np.linalg.norm(example.vel0,axis=3).max()}')
          
    start_time = time.time()
```
